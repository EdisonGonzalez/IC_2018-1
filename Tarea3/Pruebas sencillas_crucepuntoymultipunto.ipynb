### Realización de cruce multipunto:

```python
a = np.array([3,2,2,4,4,2,1,3]) #Padre1
b = np.array([4,4,1,3,4,1,1,2]) #Padre2
def CruceMultipunto(a,b):
    cruce1 = np.random.randint(0,len(a))#Primer punto de cruce
    cruce2 = 1+np.random.randint(cruce1,len(a))#Se le suma 1 para que el segundo cruce se haga al menos una posicion más allá 
    print("corte1: ",cruce1)
    print("corte2: ",cruce2)
    temp = np.zeros(len(a))
    print("cromosoma a sin cruzar: ",a)
    print("cromosoma b sin cruzar: ",b)
    temp[cruce1:cruce2] = b[cruce1:cruce2]
    b[cruce1:cruce2] = a[cruce1:cruce2]
    a[cruce1:cruce2]=temp[cruce1:cruce2]

    #print("cromosoma a cruzado: ",a)
    #print("cromosoma b cruzado: ",b)
    return a,b
child1 ,child2 = CruceMultipunto(a,b) 
print("a Luego de la funcion de cruce de un punto: ",child1)
print("b Luego de la funcion de cruce de un punto: ",child2)
```

### Realización de cruce de un punto:
En este cruce de punto se selecciona un cruce de punto aleatorio y las colas de sus padres se intercambian para obtener nuevos descendientes. 

```python
a = np.array([3,2,2,4,4,2,1,3])
b = np.array([4,4,1,3,4,1,1,2])
def CruceDeUnPunto(a,b):
    cruce1 = np.random.randint(0,len(a))#Primer punto de cruce
    #cruce2 = 1+np.random.randint(cruce1,len(a))#Se le suma 1 para que el segundo cruce se haga al menos una posicion más allá 
    print("corte1: ",cruce1)
    temp = np.zeros(len(a))
    print("cromosoma a sin cruzar: ",a)
    print("cromosoma b sin cruzar: ",b)
    temp[cruce1:len(a)] = b[cruce1:len(a)]
    b[cruce1:len(a)] = a[cruce1:len(a)]
    a[cruce1:len(a)]=temp[cruce1:len(a)]

    return a,b
child1 ,child2 = CruceDeUnPunto(a,b) 
print("a cruzado Luego de la funcion: ",child1)
print("b cruzado de la funcion: ",child2)
```

### Mutacion
En términos simples, la mutación se puede definir como un pequeño ajuste aleatorio en el
cromosoma, para obtener una nueva solución.
Se usa para mantener e introducir diversidad en la población genética y generalmente se
aplica con una baja probabilidad - pm.
La mutación es la parte de la GA que está relacionada con la "exploración" del espacio de
búsqueda. Se ha observado que la mutación es esencial para la convergencia de la GA,
mientras que el cruce no lo es.

### Random Resetting:
Es una extensión del BIt flip para la representación entera. En esta, se asigna un valor
aleatorio del conjunto de valores permisibles a un gen elegido al azar.

```python
K = 4#Clusters
a = np.array([3,2,2,4,4,2,1,3]) # Cromosoma a mutar
print("cromosoma 'a' Sin mutar: ",a)
def Mutacion_RandomResetting(a,K):
    gen1_elegido=np.random.randint(0,len(a)) #Se escoge el gen para mutar aleatoriamente 
    print("gen elegido para mutar: ",gen1_elegido)
    resetting=np.random.randint(0,K)#mutación
    print("Resetting: ",resetting)
    a[gen1_elegido]=resetting
    #print("cromosoma 'a' mutado: ",a)
    return a
mutacion = Mutacion_RandomResetting(a,K) 
print("Cromosoma mutado: ",mutacion)
```
